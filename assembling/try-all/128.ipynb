```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
pic = cv2.imread('/home/tsdhrm/Desktop/moiOneHundred-20220915T100556Z-001/moiOneHundred/138.jpg')
```

```python
new_pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
roi = new_pic[700:1400, 400:2500]
plt.imshow(roi)
```

```python
def adjust_gamma(image, gamma=1.0):
    table = np.array([((i / 255.0) ** gamma) * 255
      for i in np.arange(0, 256)]).astype("uint8")

    return cv2.LUT(image, table)
```

```python
gray = cv2.cvtColor(roi, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray, 'gray')
```

```python
new_roi = roi[10:500, 300:1300]
blurred = cv2.medianBlur(new_roi, 45)
```

```python
plt.imshow(blurred, 'gray')
```

```python
gray = cv2.cvtColor(blurred, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(gray, 'gray')
```

```python
dilated = cv2.dilate(gray, kernel=cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (100,100)))
```

```python
plt.imshow(dilated, 'gray')
```
