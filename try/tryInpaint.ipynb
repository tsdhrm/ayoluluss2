```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_dir = r'/home/tsdhrm/Pictures/thesis/dataset/100DataNew/negative/drishtiGS_101.png'
IDEAL = r'/home/tsdhrm/Pictures/thesis/dataset/100DataNew/positive/drishtiGS_069.png'
pos_dir = r'/home/tsdhrm/Pictures/thesis/dataset/100DataNew/positive/drishtiGS_068.png'
```

```python
img = cv2.imread(image_dir)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ideal = cv2.imread(IDEAL)
ideal = cv2.cvtColor(ideal, cv2.COLOR_BGR2RGB)
pos = cv2.imread(pos_dir)
pos = cv2.cvtColor(pos, cv2.COLOR_BGR2RGB)
```

```python
from skimage.exposure import match_histograms
```

```python
neg_matched = match_histograms(img, ideal, multichannel=True)
pos_matched = match_histograms(pos, ideal, multichannel=True)
```
