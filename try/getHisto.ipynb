# Histogram of all images

```python
import cv2
import matplotlib.pyplot as plt
import os
```

```python
def display_all_histo(imgs, rowcols=(12,7), size=(25,30)):
    fig, axs = plt.subplots(rowcols[0], rowcols[1], figsize=size)
    k = 0
    for i in range(rowcols[0]):
        for j in range(rowcols[1]):
            color = ('b', 'g', 'r')
            for i, col in enumerate(color):
                histr = cv2.calcHist([list(imgs.values())[k]], [i], None, [256], [0,256])
                axs[i, j].plot(histr, color=col)
                plt.xlim([0, 256])
            # axs[i, j].imshow()
            axs[i, j].set_title(list(imgs.keys())[k])
            k+=1
    
```

```python
def read_files(ddir):
    images = dict()
    ACCEPTED_FORMAT = ['jpg', 'png']
    for d in sorted(os.listdir(ddir)):
        if d.split('.')[1].lower() in ACCEPTED_FORMAT:
            fname = d.split('.')[0]
            print(f'reading {os.path.join(ddir, d)}...')
            images[fname] = cv2.imread(os.path.join(ddir, d))
            images[fname] = cv2.cvtColor(images[fname], cv2.COLOR_BGR2RGB)
    return images
```

```python
NEGATIVE = r'/home/tsdhrm/Pictures/thesis/dataset/100DataNew/negative'
```

```python
neg_f = read_files(NEGATIVE)
```

```python
display_all_histo(neg_f)
```
