```python
resized = {}
for img in retina:
    scale_percent = 60 # percent of original size
    width = int(retina[img].shape[1] * scale_percent / 100)
    height = int(retina[img].shape[0] * scale_percent / 100)
    dim = (width, height)
    resized[img] = cv2.resize(retina[img], dim, interpolation = cv2.INTER_AREA)
    
```

```python
def display_all(imgs, subs=(10,10), use_gray=False):
    fig, axs = plt.subplots(subs[0], subs[1], figsize=(18, 21))
    k = 0
    for i in range(subs[0]):
        for j in range(subs[1]):
            axs[i, j].imshow(list(imgs.values())[k], 'gray' if use_gray else None)
            axs[i, j].set_title(list(imgs.keys())[k])
            axs[i, j].axis('off')
            k+=1
            
def display_all_with(imgs, cc, subs=(10,10), use_gray=True):
    fig, axs = plt.subplots(subs[0], subs[1], figsize=(18, 21))
    k = 0
    for i in range(subs[0]):
        for j in range(subs[1]):
            axs[i, j].imshow(list(imgs.values())[k][:,:,cc], cmap='gray' if use_gray else None)
            axs[i, j].set_title(list(imgs.keys())[k])
            axs[i, j].axis('off')
            k+=1
            
def adjust_gamma(image, gamma=1.0):
    table = np.array([((i / 255.0) ** gamma) * 255
      for i in np.arange(0, 256)]).astype("uint8")

    return cv2.LUT(image, table)
```

```python
def preprocess(imgs, value):
    preprocessed = {}
    clahe = cv2.createCLAHE(2.0, (85,85))
    for img in imgs:
        cropped = imgs[img][450:1450, 500:2500]
        lab = cv2.cvtColor(cropped, cv2.COLOR_RGB2LAB)
        lab[:,:,0] = clahe.apply(lab[:,:,0])
        rgb = cv2.cvtColor(lab, cv2.COLOR_LAB2RGB)
        hsv = cv2.cvtColor(rgb, cv2.COLOR_RGB2HSV)
        h, s, v = cv2.split(hsv)
        lim = 255 - value
        v[v > lim] = 255
        v[v <= lim] -= value
        modified_hsv = cv2.merge([h,s,v])
        new_rgb = cv2.cvtColor(modified_hsv, cv2.COLOR_HSV2RGB)
        gray =  cv2.cvtColor(new_rgb, cv2.COLOR_RGB2GRAY)
        blurred = cv2.GaussianBlur(gray, (55, 55), 0)
        gmc = adjust_gamma(blurred, 5)
        xmax = cv2.minMaxLoc(gmc)[3][0]
        ymax = cv2.minMaxLoc(gmc)[3][1]
        coord1 = (xmax-135, ymax-135)
        coord2 = (xmax+135, ymax+135)
        color = (255,0,0)
        t = 5
        
        copied = np.copy(cropped)
        preprocessed[img] = cv2.rectangle(copied, coord1, coord2, color, t)
        
        # preprocessed[img] = cv2.medianBlur(gray, 75)
        # ret, preprocessed[img] = cv2.threshold(gmc, 128, 255, cv2.THRESH_BINARY)
                
    # display_all_with(preprocessed, 2, (10,10), True)
    display_all(preprocessed, (10,10), True)
```
