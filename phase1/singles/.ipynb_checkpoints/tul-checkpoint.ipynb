```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
pic = cv2.imread('/Users/teguhsatya/Desktop/moiOneHundred/286.jpg')
```

```python
pic = cv2.cvtColor(pic, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(pic)
```

```python
roi = pic[700:1115, 900:1400]
plt.imshow(roi)
```

```python
converted = cv2.cvtColor(roi, cv2.COLOR_RGB2XYZ)
plt.imshow(converted[:,:,2], 'gray')
```

```python
blurred = cv2.medianBlur(converted[:,:,2], 35)
```

```python
plt.imshow(blurred, 'gray')
```

```python
ret2,th2 = cv2.threshold(blurred,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
plt.imshow(th2)
```
