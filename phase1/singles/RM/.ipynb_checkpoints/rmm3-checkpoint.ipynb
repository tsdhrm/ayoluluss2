# RM Mark 3

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
rm = cv2.imread('/Users/teguhsatya/Dev/segeralulus/samples/s_rm.png')
rm_rgb = cv2.cvtColor(rm, cv2.COLOR_BGR2RGB)
rm_res = cv2.resize(rm_rgb, (int(rm_rgb.shape[1]*50/100), int(rm_rgb.shape[0]*50/100)), cv2.INTER_AREA)
```

```python
plt.imshow(rm_res)
```

```python
green = rm_res[:,:,1]
```

```python
plt.imshow(green, cmap='gray')
```

```python
clahe = cv2.createCLAHE(2, (8,8))
greencl = clahe.apply(green.copy())
```

```python
plt.imshow(greencl, 'gray')
```

```python
plt.imshow(green, 'gray')
```

```python
greenbl = cv2.GaussianBlur(green.copy(), (21,21),0)
```

```python
plt.imshow(greenbl, 'gray')
```

```python
opened = cv2.morphologyEx(greenbl.copy(), cv2.MORPH_OPEN, kernel=np.ones((15,15), np.uint8))
```

```python
plt.imshow(opened, 'gray')
```

```python
dilate = cv2.dilate(greenbl.copy(), kernel=np.ones((15,15), np.uint8)) # was 19,19
```

```python
plt.imshow(dilate, 'gray')
```

```python
ret3,th3 = cv2.threshold(dilate.copy(),0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
plt.imshow(th3, cmap='gray')
```

```python
contours, hierarchy = cv2.findContours(th3.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
thh = th3.copy()
for contour in contours:
    ellipse = cv2.fitEllipse(contour)
    cv2.ellipse(thh, ellipse, (255, 0, 0), 1, cv2.LINE_AA)
    plt.imshow(thh, 'gray')
```

```python
thh = th3.copy()
for contour in contours:
    ellipse = cv2.fitEllipse(contour)
    od = np.zeros_like(thh, dtype='uint8')
    cv2.ellipse(od, ellipse, (255, 255, 255), -1)
    plt.imshow(od, 'gray')
```

```python
plt.imshow(od, cmap='gray')
```

# Alternates

```python
erosion = cv2.erode(dilate.copy(), kernel=np.ones((10,10), np.uint8))
```

```python
plt.imshow(erosion, 'gray')
```

```python
ret4,th4 = cv2.threshold(erosion.copy(),0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
plt.imshow(th4)
```

```python
plt.imshow(rm_res)
```

```python
cont, hcy = cv2.findContours(th4.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
thhh = th4.copy()
roic = rm_res.copy()
for contour in cont:
    ellipse = cv2.fitEllipse(contour)
    od = np.zeros_like(thh, dtype='uint8')
    cv2.ellipse(od, ellipse, (255, 255, 255), -1)
    cv2.ellipse(roic, ellipse, (0, 255, 0), 1, cv2.LINE_AA)
    plt.imshow(od, 'gray')
```
