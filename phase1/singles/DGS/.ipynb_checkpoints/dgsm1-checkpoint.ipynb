```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
dgs = cv2.imread('/Users/teguhsatya/Dev/segeralulus/samples/s_drishti.png')
```

```python
dgs_rgb = cv2.cvtColor(dgs, cv2.COLOR_BGR2RGB)
```

```python
dgs_res = cv2.resize(dgs_rgb, (int(dgs_rgb.shape[1]*30/100), int(dgs_rgb.shape[0]*30/100)), cv2.INTER_AREA)
```

```python
dgs_res.shape
```

```python
plt.imshow(dgs_res)
```

```python
plt.imshow(dgs_res[:,:,0], cmap='gray')
```

```python
plt.imshow(dgs_res[:,:,1], cmap='gray')
```

```python
plt.imshow(dgs_res[:,:,2], cmap='gray')
```

```python
# median filter for red
dgs_median = cv2.medianBlur(dgs_res[:,:,0].copy(), 7)
```

```python
plt.imshow(dgs_median, cmap='gray')
```

```python
# clahe
clahe = cv2.createCLAHE(1.5, (8,8))
```

```python
cl_dgs = clahe.apply(dgs_median.copy())
```

```python
plt.imshow(cl_dgs, cmap='gray')
```

```python
hsv_dgs = cv2.cvtColor(dgs_res.copy(), cv2.COLOR_RGB2HSV)
```

```python
plt.imshow(hsv_dgs[:,:,2], cmap='gray')
```

```python
h_clahe = clahe.apply(hsv_dgs[:,:,2])
```

```python
plt.imshow(h_clahe, 'gray')
```

```python
h_median = cv2.medianBlur(h_clahe.copy(), 25)
```

```python
plt.imshow(h_median, 'gray')
```

```python
hh_med = h_median.copy()
xmax = cv2.minMaxLoc(hh_med)[3][0]
ymax = cv2.minMaxLoc(hh_med)[3][1]
print(xmax, ymax)
roi = hh_med[ymax-85:ymax+85,xmax-85:xmax+85]
```

```python
plt.imshow(roi, 'gray')
```

```python
roi.dtype
```

```python
rows = roi.shape[0]
circles = cv2.HoughCircles(roi.copy(), cv2.HOUGH_GRADIENT, 1, rows / 8, param1=160, param2=190, minRadius=1, maxRadius=30)
```

```python
print(circles)
```

```python
# edge detection
from skimage import feature
from skimage.transform import hough_ellipse
```

```python
roi[100, 25]
```

```python
canned = feature.canny(roi.copy(), sigma=3, low_threshold=1.5, high_threshold=2.5)
```

```python
plt.imshow(canned, cmap='gray')
```

```python
plt.hist(roi.ravel(),256,[0,256]); plt.show()
```

```python
result = hough_ellipse(canned, accuracy=20, threshold=150, min_size=100, max_size=120)
if len(result):
    print('Hadeeeeh')
```
