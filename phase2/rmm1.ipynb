```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
rm = cv2.imread('/Users/teguhsatya/Dev/segeralulus/samples/s_rm.png')
rm_rgb = cv2.cvtColor(rm, cv2.COLOR_BGR2RGB)
rm_res = cv2.resize(rm_rgb, (int(rm_rgb.shape[1]*50/100), int(rm_rgb.shape[0]*50/100)), cv2.INTER_AREA)
roi = rm_res.copy()
```

```python
plt.imshow(roi)
```

```python
def adjust_gamma(image, gamma=1.0):
    table = np.array([((i / 255.0) ** gamma) * 255
      for i in np.arange(0, 256)]).astype("uint8")

    return cv2.LUT(image, table)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (17,17))
dilate = cv2.dilate(roi.copy(), kernel)
plt.imshow(dilate)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
erode = cv2.erode(dilate.copy(), kernel)
```

```python
plt.imshow(erode)
```

```python
erogreen = erode[:,:,1].copy()
```

```python
plt.imshow(erogreen, 'gray')
```

```python
eroblur = cv2.GaussianBlur(erogreen.copy(), (15,15), 0)
```

```python
erogama = adjust_gamma(eroblur.copy(), 7)
plt.imshow(erogama, 'gray')
```

```python
ret2,th2 = cv2.threshold(erogama.copy(),0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
plt.imshow(th2, 'gray')
```

```python
cont, hcy = cv2.findContours(th2.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
```

```python
roic = roi.copy()
thhh = th2.copy()
if len(cont) > 1:
    conts = np.array([c.shape[0] for c in cont])
    cont_idx = np.argmax(conts)
    ellipse = cv2.fitEllipse(cont[cont_idx])
else:
    ellipse = cv2.fitEllipse(cont)
    
od = np.zeros_like(thhh, dtype='uint8')
cv2.ellipse(od, ellipse, (255, 255, 255), -1)
cv2.ellipse(roic, ellipse, (0, 255, 0), 1, cv2.LINE_AA)
plt.imshow(od, 'gray')  
```
